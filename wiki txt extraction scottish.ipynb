```python
wikipedia = mediawiki.MediaWiki()
```

```python
#устанавливаем язык
wikipedia.language = 'gd'
```

```python
file = 'gdwiki-20240420-pages-meta-current.csv'
all_articles = pd.read_csv(file, on_bad_lines='skip')
all_articles.head()
```

```python
id_list = all_articles['page_id'][4:]
```

```python
#Создаем файл, куда будем записывать текст, и шапку для него
import csv
with open('gdwiki-20240420-pages-meta-current.csv','w') as f:
    writer = csv.writer(f)
    writer.writerow(['Page Id', 'Article', 'Summary'])
with open('gd_dataset.txt','w', encoding="utf-8") as f:
            f.write("")
```

```python
for i in id_list:
    try:
        content = wikipedia.page(pageid=i).content
        #Пишем в текстовый файл
        with open('gd_dataset.txt','a', encoding="utf-8") as f:
            f.write(content)
        print(f"Article {i} was added")
    except:
        print(f"Article {i} was skipped")
        continue
```
