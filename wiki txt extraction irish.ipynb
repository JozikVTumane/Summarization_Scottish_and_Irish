```python
import mediawiki
import pandas as pd
```

```python
wikipedia = mediawiki.MediaWiki()
```

```python
#устанавливаем язык
wikipedia.language = 'ga'
```

```python
all_articles = pd.read_csv('gawiki.csv', on_bad_lines='skip')
all_articles.head()
```

```python
id_list = all_articles['page_id'][4:]
```

```python
#Создаем файл, куда будем записывать текст, и шапку для него
with open('gawiki_content.txt','w', encoding="utf-8") as f:
            f.write()
```

```python
for i in id_list:
    try:
        content = wikipedia.page(pageid=i).content
        summary = wikipedia.page(pageid=i).summary
        #Пишем в текстовый файл
        with open('gawiki_content.txt','a', encoding="utf-8") as f:
            f.write(content)
        print(f"Article {i} was added")
    except:
        print(f"Article {i} was skipped")
        continue
```
